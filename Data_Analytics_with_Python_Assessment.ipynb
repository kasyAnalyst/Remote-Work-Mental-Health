## Shopping Cart System with Discounts

Write a Python program to simulate a shopping cart system for an online store. The program will calculate the total cost of items, apply discounts, and check if the total exceeds a specified budget.

1.	Variables and Lists:
  - Define a `budget` variable with an initial value of 200.
  
  - Create two empty lists called `item_names_list` and `item_prices_list` to store the name and price of each item separately.

```python
# Define the budget with an initial value
budget = 200

# Create empty lists to store item names and prices
item_names_list = []
item_prices_list = []


```

2. Functions:
  - Write a function `add_item_to_cart(item_name, item_price)` that takes the itemâ€™s name and price as arguments, appends the name to item_names and the price to item_prices, and returns both updated lists.
  
  - Write a function `calculate_total(item_prices)` that calculates and returns the total cost of all items in item_prices.

    Conditions:
    - If the total cost exceeds the budget after adding an item, print "Budget exceeded!" and stop adding more items.
    - If the total cost is within budget and exceeds $100, apply a 10% discount on the total and print the discounted total.

```python
# Function to add an item to the cart
def add_item_to_cart(item_name, item_price):
    # Append the item name and price to the respective lists
    item_names_list.append(item_name)
    item_prices_list.append(item_price)
    return item_names_list, item_prices_list

# Function to calculate the total cost of items in the cart
def calculate_total(item_prices):
    return sum(item_prices)

# Function to apply discount and check budget
def apply_discount_and_check_budget(total, budget):
    if total > budget:
        print("Budget exceeded!")
        return None  # Stop adding more items if budget is exceeded
    elif total > 100:
        discount = 0.1 * total  # 10% discount
        discounted_total = total - discount
        print(f"Total exceeds $100, applying 10% discount: ${discounted_total:.2f}")
        return discounted_total
    else:
        print(f"Total within budget: ${total:.2f}")
        return total
```

3.	Loop and Input:
  - Start the input only once the user says 'start'
  - Use a loop to allow the user to add items to the cart by entering an item name and price. The loop should stop when the user types 'done'.
  - For each item, add it to item_names and item_prices using add_item_to_cart, then update the total cost using calculate_total.

Output:
  - After the loop ends, display the final cart with each item and its price, the initial total, any applicable discount, and the final total.


```python
# Start the input only when the user types 'start'
def start_input():
    user_input = input("Type 'start' to begin adding items to the cart: ").strip().lower()
    if user_input == 'start':
        while True:
            # Ask user for item name and price
            item_name = input("Enter item name (or 'done' to finish): ").strip()
            if item_name.lower() == 'done':
                break
            try:
                item_price = float(input(f"Enter price for '{item_name}': "))
                # Add item to the cart
                add_item_to_cart(item_name, item_price)
                
                # Calculate the total cost
                total = calculate_total(item_prices_list)
                
                # Check the budget and apply discount if necessary
                final_total = apply_discount_and_check_budget(total, budget)
                if final_total is None:
                    break  # Stop adding more items if budget exceeded
                
            except ValueError:
                print("Invalid price entered. Please enter a valid number for the price.")
```

```python
# After the loop ends, display the final cart and results
def display_final_cart():
    print("\nShopping Cart Summary:")
    print("Items in Cart:")
    for i in range(len(item_names_list)):
        print(f"{item_names_list[i]}: ${item_prices_list[i]:.2f}")
    
    total = calculate_total(item_prices_list)
    final_total = apply_discount_and_check_budget(total, budget)
    
    print(f"\nInitial Total: ${total:.2f}")
    if final_total is not None:
        print(f"Final Total after discount (if applicable): ${final_total:.2f}")
    else:
        print("No further items added due to budget exceeding.")
```

```python

# Step 1: Filter 'Stress_Level' to only 'low' and 'high'
df['Stress_Level'] = df['Stress_Level'].str.strip().str.lower()
df = df[df['Stress_Level'].isin(['low', 'high'])]

# Map stress levels to numeric values
stress_mapping = {'low': 1, 'high': 3}
df['Stress_Level_Num'] = df['Stress_Level'].map(stress_mapping)

# Step 2: Now, create a new column to categorize stress levels as 'Low' or 'High'
# In the mapping, 'low' is 1, and 'high' is 3, so I categorize accordingly
df['Stress_Category'] = df['Stress_Level_Num'].apply(lambda x: 'High' if x == 3 else 'Low')

# Step 3: Group by Job_Role and Stress_Category, then calculate the mean
stress_avg_by_role = df.groupby(['Job_Role', 'Stress_Category'])['Stress_Level_Num'].mean().unstack()

# Step 4: Plot
stress_avg_by_role.plot(kind='bar', figsize=(10, 6), color=['lightblue', 'lightcoral'])
plt.title('Average Stress Level by Job Role (High vs Low Stress)', fontsize=14)
plt.xlabel('Job Role', fontsize=12)
plt.ylabel('Average Stress Level', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Stress Level', loc='upper right', labels=['Low', 'High'])
plt.tight_layout()
plt.show()

```
